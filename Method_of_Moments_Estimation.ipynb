<h3>Solution

To estimate the parameter $\lambda$ of a Poisson distribution using the method of moments, we proceed with the following steps:

**Step 1: Define the Moment of the Poisson Distribution**

The Poisson distribution has one parameter, $\lambda$, which represents both the mean and the variance of the distribution.

For the Poisson distribution:
- The first moment (mean) $\mu_1$ is equal to $\lambda$.

So, for the Poisson distribution:

$$
\mu_1 = \mathbb{E}(X) = \lambda
$$

**Step 2: Theoretical Moment for the Data**

Now, use the first moment to estimate $\lambda$. The first theoretical moment $\hat{\mu}_1$, which is the sample mean, is given by:

$$
\hat{\mu}_1 = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

where $x_i$ are the observed values (sample data) and $N$ is the total number of observations.

**Step 3: Set the Sample Moment Equal to the Population Moment**

Using the method of moments, we equate the theoretical moment $\hat{\lambda}$ (the sample mean) to the population moment $\lambda$.

$$
\hat{\mu}_1 = \lambda
$$

Thus, the method of moments estimator for $\lambda$ is simply the sample mean:

$$
\hat{\lambda} = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

**Final Estimate**

The estimated value of $\lambda$, denoted by $\hat{\lambda}$, is the average of the observed data points:

$$
\hat{\lambda} = \frac{1}{N} \sum_{i=1}^{N} x_i
$$

````r
# Example R code to estimate lambda for a Poisson distribution

# Generating some sample data that follows a Poisson distribution
set.seed(123) # Setting seed for reproducibility
observed_data <- rpois(100, lambda = 4) # 100 observations with true lambda = 4

# Step 1: Calculate the sample mean (method of moments estimator for lambda)
lambda_hat <- mean(observed_data)

# Print the estimated value of lambda
cat("Estimated lambda (λ):", lambda_hat, "\n")

# Step 2: Compare the true lambda with the estimated one
true_lambda <- 4
cat("True lambda (λ):", true_lambda, "\n")
````

````r
# Example R code to estimate the mean (μ) and variance (σ^2) of normally distributed data

# Step 1: Generate some sample data that follows a normal distribution
set.seed(123) # Setting seed for reproducibility
observed_data <- rnorm(100, mean = 5, sd = 2) # 100 observations with mean = 5 and sd = 2

# Step 2: Method of moments estimator for the mean (μ)
mu_hat <- mean(observed_data)

# Step 3: Method of moments estimator for the variance (σ^2)
sigma_squared_hat <- mean((observed_data - mu_hat)^2)

# Print the estimated values of μ and σ^2
cat("Estimated mean (μ):", mu_hat, "\n")
cat("Estimated variance (σ^2):", sigma_squared_hat, "\n")

# Step 4: Compare with the true parameters
true_mu <- 5
true_sigma_squared <- 2^2
cat("True mean (μ):", true_mu, "\n")
cat("True variance (σ^2):", true_sigma_squared, "\n")
````

**First**, we know the following formulas for mean and variance for lognormal distribution:
$$\mu_1 = e^{(\mu \space + \space 0.5 \space \sigma^{2})}$$

**Second**, we need to get $\mu$ and $\sigma^{2}$ in terms of $\mu_1$ and $\mu_2$ as the following:

$$\log\left( \mu_1 \right) = \mu \space + \space 0.5 \space \sigma^{2} \space\space\space\space\space\space\space\space\space\space\space\space eq(1)$$
$$\log\left( \mu_2 \right) = 2 \mu \space + \space 2 \sigma^{2} \space\space\space\space\space\space\space\space\space\space\space\space\space\space eq(2)$$

$$-4 \times eq(1) + eq(2)$$
$$-2 \mu = -4 \log\left( \mu_1 \right) + \log\left( \mu_2 \right) \space\space\space\space (÷ -2)$$
$$\mu = 2 \log\left( \mu_1 \right) - \frac{1}{2} \log\left( \mu_2 \right)$$

$$\mu = \frac{1}{2} \log\left(\frac{{\mu_1^4}}{{\mu_2}}\right)$$

$$-2 \times eq(1) + eq(2)$$
$$\sigma^2 = \log\left( \mu_2 \right) - 2 \log\left( \mu_1 \right)$$

$$\sigma^{2} = \log\left(\frac{{\mu_2}}{{\mu_1^2}}\right)$$

**Third**, using the theoretical moments to obtain the final formulas:

$$\mu_1 = \frac{1}{N} \sum_{1}^{N} x_{i}$$
$$\mu_2 = \frac{1}{N} \sum_{1}^{N} x_{i}^{2}$$


**Finally:**

$$\hat{\mu} = \frac{1}{2} \log\left(\frac{{\mu_1^4}}{{\mu_2}}\right) = \frac{1}{2} \log\left(\frac{({\frac{1}{N} \sum_{1}^{N} x_{i}})^4}{{\frac{1}{N} \sum_{1}^{N} x_{i}^{2}}}\right)$$
$$\hat{\sigma}^{2} = \log\left(\frac{{\mu_2}}{{\mu_1^2}}\right) = \log\left(\frac{{\frac{1}{N} \sum_{1}^{N} x_{i}^{2}}}{{(\frac{1}{N} \sum_{1}^{N} x_{i}})^2}\right)$$

Write two functions `estimate_mu(X)`, and `estimate_sigma_sq(X)` where the input **X** is a sample of N data points stored in a list or an array. The functions should implement the estimator equations you derived in question 1, and return the estimated $\hat{μ}$, or $\hat{σ^2}$ respectively.

````r
calculate_moments <- function(X) {
  # Calculate the first and second moments of the data X

  n <- length(X)
  m1 <- 1 / n * sum(X)
  m2 <- 1 / n * sum(X^2)

  return(list(m1 = m1, m2 = m2))
}
````

````r
estimate_mu <- function(X) {
  # Estimate the mu parameter based on the data X

  moments <- calculate_moments(X)
  m1 <- moments$m1
  m2 <- moments$m2
  estimated_mu <- 0.5 * log((m1^4) / m2)

  return(estimated_mu)
}
````

````r
estimate_sigma_sq <- function(X) {
  # Estimate the sigma squared parameter based on the data X

  moments <- calculate_moments(X)
  m1 <- moments$m1
  m2 <- moments$m2
  estimated_sigma_sq <- log(m2 / (m1^2))

  return(estimated_sigma_sq)
}
````

Assuming the true parameters are $μ=1$, and $σ^2 = 0.25$, conduct a repeated sampling simulation (monte carlo simulation) with **150** repititions (num_samples = 150) and varying sample sizes N, drawn from the lognormal distribution $Lognorm(1, 0.25)$. Plot the sampling distributions of sample $\hat{\mu}$, and $\hat{σ^2}$ estimates for sample sizes N=3, N=10, N=50, N=100, N=1000, and N=10000.  This is very similar to what we did in lab 3.

- Based on your observation, what happens to the sampling distributions of $\hat{\mu}$, and $\hat{σ^2}$ when varying the sample size N?

````r
sample_distribution <- function(sample_size_N, estimate_sigma_sq_function) {
  # Generate samples from a distribution and estimate parameters

  estimated_mu_array <- c()
  estimated_sigma_sq_array <- c()

  for (i in 1:150) {
    mu <- 1
    sigma <- sqrt(0.25)

    # Generate log-normal random samples
    X <- rlnorm(sample_size_N, meanlog = mu, sdlog = sigma)

    # Estimate mu and append to array
    estimated_mu <- estimate_mu(X)
    estimated_mu_array <- c(estimated_mu_array, estimated_mu)

    # Estimate sigma squared and append to array
    estimated_sigma_sq <- estimate_sigma_sq_function(X)
    estimated_sigma_sq_array <- c(estimated_sigma_sq_array, estimated_sigma_sq)
  }

  return(list(estimated_mu_array = estimated_mu_array, estimated_sigma_sq_array = estimated_sigma_sq_array))
}
````

````r
plot_histograms <- function(array1, array2, N, xlabel1 = 'Array 1', xlabel2 = 'Array 2',
                            bins = 30, edgecolor = 'black', range_min = 0, range_max = 2,
                            main_title = 'Histogram of Estimated Values (N = %d)') {

  # Adjust margins to make space for the centered title
  par(mfrow = c(1, 2), mar = c(5, 5, 4, 2), oma = c(0, 0, 4, 0))  # Increase outer top margin for title

  # Plot the first histogram
  hist(array1, breaks = bins, col = edgecolor, xlim = c(range_min, range_max),
       main = "", xlab = xlabel1, prob = TRUE)

  # Calculate statistics for array1
  mean1 <- mean(array1)
  variance1 <- var(array1)

  # Annotate statistics on the first histogram
  text(x = range_max * 0.7, y = max(hist(array1, breaks = bins, plot = FALSE)$density) * 0.9,
       labels = sprintf("mu = %.4f\nsigma^2 = %.4f", mean1, variance1), adj = c(0.5, 0.5), cex = 0.8)

  # Plot the second histogram
  hist(array2, breaks = bins, col = edgecolor, xlim = c(range_min, range_max),
       main = "", xlab = xlabel2, prob = TRUE)

  # Calculate statistics for array2
  mean2 <- mean(array2)
  variance2 <- var(array2)

  # Annotate statistics on the second histogram
  text(x = range_max * 0.7, y = max(hist(array2, breaks = bins, plot = FALSE)$density) * 0.9,
       labels = sprintf("mu = %.4f\nsigma^2 = %.4f", mean2, variance2), adj = c(0.5, 0.5), cex = 0.8)

  # Add a single centered title for both plots
  mtext(sprintf(main_title, N), side = 3, line = 2, outer = TRUE, cex = 1.5, font = 2)

  # Reset plot layout
  par(mfrow = c(1, 1))
}
````

````r
sample_sizes_N <- c(3, 10, 50, 100, 1000, 10000)

for (sample_size_N in sample_sizes_N) {
  # Call the sample_distribution function to get estimated mu and sigma squared arrays
  results <- sample_distribution(sample_size_N, estimate_sigma_sq)

  # Extract the estimated_mu_array and estimated_sigma_sq_array from the results
  estimated_mu_array <- results$estimated_mu_array
  estimated_sigma_sq_array <- results$estimated_sigma_sq_array

  # Plot histograms for the estimated mu and estimated sigma squared
  plot_histograms(estimated_mu_array, estimated_sigma_sq_array, sample_size_N,
                  xlabel1 = 'Estimated Mu', xlabel2 = 'Estimated Sigma Squared')
}
````

A possible different estimator for $σ^2$ is:
<h2><center>$\hat{σ^2} = \ln {( \frac{\bar{X}}{x_m \bar{M_2}}+1)}$ </center>

where $\bar{X}$ is the sample mean, $\bar{M_2}$ the sample variance (second central moment), and $x_m$ is the sample maximum.

* Implement this new estimator.
* Repeat the same monte carlo simulation you did in Q3 using this new estimator.
*  compare the performance of this estimator to the one you derived using MoM in terms of : Bias, Variance, and consistency.

````r
another_estimator_of_sigma_sq <- function(X) {
  # Estimate sigma squared using another method

  xm <- max(X)
  X_ <- mean(X)
  M2_ <- var(X)

  estimated_sigma_sq <- log((X_ / (xm * M2_)) + 1)

  return(estimated_sigma_sq)
}
````

````r
sample_sizes_N <- c(3, 10, 50, 100, 1000, 10000)

for (sample_size_N in sample_sizes_N) {
  # Call the sample_distribution function to get estimated mu and sigma squared arrays using another_estimator_of_sigma_sq
  results <- sample_distribution(sample_size_N, another_estimator_of_sigma_sq)

  # Extract the estimated_mu_array and estimated_another_sigma_sq_array from the results
  estimated_mu_array <- results$estimated_mu_array
  estimated_another_sigma_sq_array <- results$estimated_sigma_sq_array

  # Plot histograms for the estimated mu and the second sigma squared
  plot_histograms(estimated_mu_array, estimated_another_sigma_sq_array, sample_size_N,
                  xlabel1 = 'Estimated Mu', xlabel2 = 'Estimated Second Sigma Squared')
}
````
