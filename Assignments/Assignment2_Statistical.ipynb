```r
lambda<-1/2
n<-10
num_samples <-1500
sample_means_n10<-replicate(num_samples, mean(rexp(n, rate = lambda)))
hist(sample_means_n10, breaks = 30, probability = TRUE, col = "blue",
     main = "Relative Frequency Histogram (n=10)",
     xlab = "Sample Mean", ylab = "Relative Frequency")
```
