#**Exploring Statistical Measures in R:**:
Statistical measures like mean, variance, standard deviation, median, and mode are crucial in data analysis. These metrics help summarize data and understand its distribution. In this lab, we’ll explore how to calculate these measures in R. The R language provides very easy methods to calculate them.


##**Mean:**
Mean or Average is a number expressing the central or typical value in a set of data

```r
# R program to get average of a list
# Taking a list of elements
list = c(2, 4, 4, 4, 5, 5, 7, 9)
# Calculating average using mean()
print(mean(list))
```

```r
list = c(2, 40, 2, 502, 177, 7, 9)
print(mean(list))
```

##**Variance:**
Variance is the sum of squares of differences between all numbers and means.

```r
# R program to get variance of a list
# Taking a list of elements
list = c(2, 4, 4, 4, 5, 5, 7, 9)
# Calculating variance using var()
print(var(list))
```

##**Standard Deviation:**
Standard Deviation is the square root of variance. It is a measure of the extent to which data varies from the mean

```r
# R program to get standard deviation of a list
# Taking a list of elements
list = c(2, 4, 4, 4, 5, 5, 7, 9)
# Calculating standard deviation using sd()
print(sd(list))
```

##**Median:**
It is the middle value of the data set. It splits the data into two halves. If the number of elements in the data set is odd then the center element is median and if it is even then the median would be the average of two central elements.

```r
# Taking a list of elements
list = c(2, 9, 5, 4, 5, 5, 7, 9, 0)
# Compute the median value
median = median(list)
print(median)
```

##**Mode:**

It is the value that has the highest frequency in the given data set. The data set may have no mode if the frequency of all data points is the same.

Also, we can have more than one mode if we encounter two or more data points having the same frequency. There is no inbuilt function for finding mode in R, so we can create our own function for finding the mode or we can use the package called modest.

###**Creating a user-defined function for finding Mode:**

There is no in-built function for finding mode in R. So let’s create a user-defined function that will return the mode of the data passed.

We will be using the table() method for this as it creates a categorical representation of data with the variable names and the frequency in the form of a table.

We will sort the list in descending order and will return the 1 value from the sorted values.

```r
# Taking a list of elements
list = c(2, 9, 5, 4, 5, 5, 7, 9, 1)
# Compute the mode value

mode = function(){
    return(sort(-table(list))[1])
}

mode()
```

# **Calculating Statistical Measures for mtcars dataset**
we will work on the column: (disp: Displacement)

```r
# Calculating average using mean()
disp_mean = mean(mtcars$disp)
cat("The mean of Displacement column: ", disp_mean, "\n")

# Calculating variance using var()
disp_variance = var(mtcars$disp)
cat("The variance of Displacement column: ", disp_variance, "\n")

# Calculating standard deviation using sd()
disp_Stand_Dev = sd(mtcars$disp)
cat("The standard deviation of Displacement column: ", disp_Stand_Dev, "\n")

# Compute the median value
disp_median = median(mtcars$disp)
cat("The median of Displacement column: ", disp_median, "\n")

# Compute the mode value
disp_mode = mfv(mtcars$disp)
cat("The mode of Displacement column: ", disp_mode, "\n")
```

# **Visualization:**
One-dimensional Plotting:

In one-dimensional plotting, we plot one variable at a time.

For example, we may plot a variable with the number of times each of its values occurred in the entire dataset (frequency). So, it is not compared to any other variable of the dataset.

These are the 4 major types of graphs that are used for One-dimensional analysis:


*   Five Point Summary
*   Box Plotting
* Histograms
* Bar Plotting

**Five Point Summary:**

So, for any particular column of the dataset, we can generate a Five-Point summary using the summary() function.

```r
summary(mtcars)
```

**Box Plotting:**

A box plot generates a rectangle that covers the area spanned by the column of the dataset.

This leads us to the concept of **Interquartile Range (IQR)**:

The IQR is a measure of statistical dispersion, or how spread out the data points are. It is the range between the 1st quartile (Q1) and the 3rd quartile (Q3) of a dataset. Essentially, it covers the middle 50% of the data. The formula to calculate the IQR is:
IQR=Q3−Q1

Q1 (1st quartile): This is the 25th percentile of the data. It represents the value below which 25% of the data points fall.

Q3 (3rd quartile): This is the 75th percentile of the data. It represents the value below which 75% of the data points fall.

```r
boxplot(mtcars$disp, col="green")
```

**Green Box:** The green area (the box) spans the interquartile range (IQR), which is the middle 50% of the data. This means that the bottom of the box corresponds to the 1st quartile (Q1 or 25th percentile), and the top of the box corresponds to the 3rd quartile (Q3 or 75th percentile).

**Median (black line inside the box):** The bold black line inside the green box represents the median (the 50th percentile) of the displacement values.
Whiskers:

**The dashed lines (whiskers):** extending from the top and bottom of the box show the range of the data, extending up to 1.5 times the IQR from Q1 and Q3. Data points beyond these whiskers are considered outliers, though in this case, there are no visible outliers.

**Histograms:**

Histograms are the most widely used plots for analyzing datasets.

Histograms are one of the most common ways to visualize the distribution of continuous data. In this case, the data being analyzed is displacement of each car in the mtcars dataset.

```r
# Histogram 1: Default number of bins
hist(mtcars$disp, col = "green",
     main = "Histogram of mtcars$disp",
     xlab = "Displacement (disp)",
     ylab = "Frequency")

# Histogram 2: Breaks set to 25
hist(mtcars$disp, col = "green", breaks = 25,
     main = "Histogram of mtcars$disp with 25 Breaks",
     xlab = "Displacement (disp)",
     ylab = "Frequency")

# Histogram 3: Breaks set to 50
hist(mtcars$disp, col = "green", breaks = 50,
     main = "Histogram of mtcars$disp with 50 Breaks",
     xlab = "Displacement (disp)",
     ylab = "Frequency")
```

The different plots with varying breaks allow you to explore the distribution at different levels of granularity:

The number of breaks determines how many bins or intervals the data will be divided into. Adjusting the breaks changes the width of each bin and how the data is displayed. Fewer breaks result in wider bins, while more breaks create narrower bins, giving a more detailed view of the data distribution.

How Breaks Work:
Fewer Breaks (Wider Bins):
This creates broader intervals, which can provide a general overview of the data distribution but might miss some of the finer details.

More Breaks (Narrower Bins):
This creates smaller intervals, showing more detail and variation in the data.

**3-D Graphs in R:**

Here we will use preps() function, This function is used to create 3D surfaces in perspective view. This function will draw perspective plots of a surface over the x–y plane.

```r
# Load the mtcars dataset
data <- mtcars

# Prepare variables from mtcars
x <- data$mpg  # Miles per gallon (mpg) for x-axis
y <- data$hp   # Horsepower (hp) for y-axis
z <- data$wt   # Weight (wt) for z-axis

# Create a grid for plotting
x_seq <- seq(min(x), max(x), length = 30)
y_seq <- seq(min(y), max(y), length = 30)
z_matrix <- outer(x_seq, y_seq, function(x, y) sqrt(x ^ 2 + y ^ 2))  # Similar transformation for z

# Plot the 3D surface using persp
persp(x_seq, y_seq, z_matrix,
      main = "3D Perspective Plot of mtcars Data",
      xlab = "Miles per Gallon (mpg)",
      ylab = "Horsepower (hp)",
      zlab = "Transformed Weight",
      theta = 30, phi = 15,
      col = "orange", shade = 0.4)
```

#**ggplot**

ggplot2 is a powerful and versatile R package for creating visualizations, offering a more flexible approach compared to the base plotting functions.

Built on the "Grammar of Graphics," ggplot2 allows you to layer components like data, aesthetics, and geometries in a structured and modular way. This enables you to create complex plots with clear and concise syntax.

Building Blocks of layers with the grammar of graphics
* Data: The element is the data set itself
* Aesthetics: The data is to map onto the Aesthetics attributes such as x-axis, y-axis, color, fill, size, labels, shape, line width, line type
* Geometrics: How our data being displayed using point, line, histogram, bar, boxplot
* Facets: It displays the subset of the data using Columns and rows
Statistics: Binning, smoothing, descriptive, intermediate
* Coordinates: the space between data and display using Cartesian, fixed, polar, limits
* Themes: Non-data link

image.png

```r
library(ggplot2)
```

**First Layer: Data**

ggplot2 in R the data Layer we define the source of the information to be visualize, let’s use the mtcars dataset in the ggplot2 package.

```r
ggplot(data = mtcars) +
labs(title = "MTCars Data Plot")
```

**Aesthetic Layer:**

Here we will display and map dataset into certain aesthetics.

```r
# Aesthetic Layer
ggplot(data = mtcars, aes(x = hp, y = mpg, col = disp))+
labs(title = "MTCars Data Plot")
```

**Geometric layer:**

geometric layer control the essential elements, see how our data being displayed using point, line, histogram, bar, boxplot.

```r
# Geometric layer
ggplot(data = mtcars, aes(x = hp, y = mpg, col = disp)) +
geom_point() +
labs(title = "Miles per Gallon vs Horsepower",
	x = "Horsepower",
	y = "Miles per Gallon")
```

In the same layer (Geometric layer), We can add Size, color, and shape and then plotting the Histogram plot

```r
# Adding size
ggplot(data = mtcars, aes(x = hp, y = mpg, size = disp)) +
geom_point() +
labs(title = "Miles per Gallon vs Horsepower",
	x = "Horsepower",
	y = "Miles per Gallon")

# Adding shape and color
ggplot(data = mtcars, aes(x = hp, y = mpg, col = factor(cyl),
	shape = factor(am))) +geom_point() +
labs(title = "Miles per Gallon vs Horsepower",
	x = "Horsepower",
	y = "Miles per Gallon")

# Histogram plot
ggplot(data = mtcars, aes(x = hp)) +
geom_histogram(binwidth = 5) +
labs(title = "Histogram of Horsepower",
	x = "Horsepower",
	y = "Count")
```

**Facet Layer:**

 facet layer is used to split the data up into subsets of the entire dataset and it allows the subsets to be visualized on the same plot. Here we separate rows according to transmission type and Separate columns according to cylinders.

```r
# Facet Layer
# Separate rows according to transmission type
p <- ggplot(data = mtcars, aes(x = hp, y = mpg, shape = factor(cyl))) + geom_point()

p + facet_grid(am ~ .) +
labs(title = "Miles per Gallon vs Horsepower",
	x = "Horsepower",
	y = "Miles per Gallon")

# Separate columns according to cylinders
p <- ggplot(data = mtcars, aes(x = hp, y = mpg, shape = factor(cyl))) + geom_point()

p + facet_grid(cyl ~ .) +
labs(title = "Miles per Gallon vs Horsepower",
	x = "Horsepower",
	y = "Miles per Gallon")
```

**Statistics layer:**

we transform our data using binning, smoothing, descriptive, intermediate

```r
ggplot(data = mtcars, aes(x = hp, y = mpg)) +
geom_point() +
stat_smooth(method = lm, col = "red") +
labs(title = "Miles per Gallon vs Horsepower")
```

**Coordinates layer:**

data coordinates are mapped together to the mentioned plane of the graphic and we adjust the axis and changes the spacing of displayed data with Control plot dimensions.

```r
ggplot(data = mtcars, aes(x = wt, y = mpg)) +
geom_point() +
stat_smooth(method = lm, col = "red") +
scale_y_continuous("Miles per Gallon", limits = c(2, 35), expand = c(0, 0)) +
scale_x_continuous("Weight", limits = c(0, 25), expand = c(0, 0)) +
coord_equal() +
labs(title = "Miles per Gallon vs Weight",
	x = "Weight",
	y = "Miles per Gallon")
```

Coord_cartesian() to proper zoom in:

```r
# Add coord_cartesian() to proper zoom in
ggplot(data = mtcars, aes(x = wt, y = hp, col = am)) +
						geom_point() + geom_smooth() +
						coord_cartesian(xlim = c(3, 6))
```

**Theme Layer:**

controls the finer points of display like the font size and background color properties.

```r
ggplot(data = mtcars, aes(x = hp, y = mpg)) +
geom_point() +
facet_grid(. ~ cyl) +
theme(plot.background = element_rect(fill = "green", colour = "gray")) +
labs(title = "Miles per Gallon vs Horsepower")
```

```r
ggplot(data = mtcars, aes(x = hp, y = mpg)) +
		geom_point() + facet_grid(am ~ cyl) +
		theme_gray()+
labs(title = "Miles per Gallon vs Horsepower")
```

```r
library(ggplot2)
library(gridExtra)

# Selecting specific columns from mtcars dataset
selected_cols <- c("mpg", "disp", "hp", "drat")
selected_data <- mtcars[, selected_cols]

# Create histograms for individual variables
hist_plot_mpg <- ggplot(selected_data, aes(x = mpg)) +
geom_histogram(binwidth = 2, fill = "blue", color = "white") +
labs(title = "Histogram: Miles per Gallon", x = "Miles per Gallon", y = "Frequency")

hist_plot_disp <- ggplot(selected_data, aes(x = disp)) +
geom_histogram(binwidth = 50, fill = "red", color = "white") +
labs(title = "Histogram: Displacement", x = "Displacement", y = "Frequency")

hist_plot_hp <- ggplot(selected_data, aes(x = hp)) +
geom_histogram(binwidth = 20, fill = "green", color = "white") +
labs(title = "Histogram: Horsepower", x = "Horsepower", y = "Frequency")

hist_plot_drat <- ggplot(selected_data, aes(x = drat)) +
geom_histogram(binwidth = 0.5, fill = "orange", color = "white") +
labs(title = "Histogram: Drat", x = "Drat", y = "Frequency")

# Arrange the plots in a grid
grid.arrange(hist_plot_mpg, hist_plot_disp, hist_plot_hp, hist_plot_drat,
			ncol = 2)
```
